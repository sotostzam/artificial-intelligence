# Gradient Descent using Python

Import the numpy library, which is usefull for scientific computing in Python
as well as the matplotlib library, used for its plotting capabilities.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Define the function to run gradient descent uppon.

```python
def f(x):
    return 2 * x * x * np.cos(x) - 5 * x
```

```python
x = np.linspace(-5,5,100)

fig = plt.figure()
#ax = fig.add_subplot(1, 1, 1)
ax = fig.add_axes([0,0,1,1])
ax.set_title("Gradient Descent")
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#ax.xaxis.set_ticks_position('bottom')
#ax.yaxis.set_ticks_position('left')
plt.grid()
plt.plot(x, f(x), 'r')
```
